## Implémentation des Piles en Python (POO)
Les classes Maillon et Pile sont définies dans la cellule suivante :

```python
class Maillonp:
    def __init__(self,valeur,suivant=None):
        self.valeur=valeur
        self.suivant=suivant
        
class Pile:
    def __init__(self):
        self.taille=0
        self.sommet=None
        
    def empiler(self,valeur):
        self.sommet=Maillonp(valeur,self.sommet)
        self.taille+=1
    
    def depiler(self):
        if(self.taille>0):
            valeur=self.sommet.valeur
            self.sommet=self.sommet.suivant
            self.taille-=1
            return valeur
        
    def estVide(self):
        return self.taille == 0
    
    def __str__(self):
        ch="\nétat de la pile\n"
        maillon=self.sommet
        while maillon != None:
            ch += str(maillon.valeur) + " "
            maillon=maillon.suivant
        return ch
```
