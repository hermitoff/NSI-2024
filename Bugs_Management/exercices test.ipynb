```python
def erreur6(a, b):
    """
    a, b – int, entiers strictement positifs
    Sortie: int – pgcd des entiers a et b
    """
    if b == 0:
        return a
    while b > 0:
        reste = a % b
        a = b
        b = reste
    return a
```
