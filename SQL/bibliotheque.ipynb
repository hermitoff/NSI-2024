```sql
CREATE TABLE IF NOT EXISTS `emprunter` (
  `numlecteur` int(11) NOT NULL,
  `numouvrage` int(11) NOT NULL,
  `dateemprunt` date NOT NULL,
  `dateretour` date DEFAULT NULL,
  PRIMARY KEY (`numlecteur`,`numouvrage`,`dateemprunt`),
  FOREIGN KEY (`numlecteur`) REFERENCES `lecteur` (`numlecteur`),
  FOREIGN KEY (`numouvrage`) REFERENCES `ouvrage` (`numouvrage`)
);
```

## R01 : Qui est le lecteur n°25 ?

```sql
SELECT numlecteur, nomlecteur, prenomlecteur
FROM lecteur
WHERE numlecteur = 25
```

## R02 : Liste des lecteurs dont le nom de famille commence par la lettre « A »

```sql
SELECT numlecteur, nomlecteur, prenomlecteur
FROM lecteur
WHERE nomlecteur LIKE "A%";
```

## R03 : Liste des ouvrages qui traitent de la méthode Merise. 

```sql
SELECT titreouvrage
FROM ouvrage
WHERE titreouvrage LIKE "%Merise%";
```

## R04 : Liste des ouvrages qui traitent du « registre » à partir de l’année 2000.

```sql
SELECT titreouvrage, anneeparution
FROM ouvrage
WHERE anneeparution > 2000
AND titreouvrage LIKE "%registre%";
```

## R05 : Liste des lecteurs qui portent le nom (ou le prénom) « BERTRAND »

```sql
SELECT numlecteur, nomlecteur, prenomlecteur
FROM lecteur
WHERE nomlecteur LIKE "%BERTRAND%"
OR prenomlecteur LIKE "%BERTRAND%";
```

## R06 : Titre des ouvrages actuellement empruntés ou ayant déjà été empruntés.

```sql
SELECT numouvrage, dateemprunt, dateretour
FROM emprunter
WHERE dateretour IS NULL;
```

## R07 : Nom des lecteurs ayant effectué au moins un emprunt depuis le début de l’année 2024. 

```sql
SELECT numlecteur, numouvrage, dateemprunt, dateretour
FROM emprunter
WHERE dateemprunt LIKE "%2024%"
```
